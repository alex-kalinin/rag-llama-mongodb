```python
from llama_cpp import Llama
import torch

class Object(object):
    pass
```

```python
t = Object()
```

```python
t.llm_path = "../../data/llama/llama-2-13b-chat.Q6_K.gguf"
t.layers = 50

# t.llm_path = "../../data/llama/llama-2-70b-chat.Q6_K.gguf"
# t.layers = 22
```

```python
t.llm = Llama(
    model_path=t.llm_path,
    n_gpu_layers=t.layers,
    n_threads=10, 
    n_ctx=8126, 
#     n_batch=512, 
    verbose=True,
)
```

```python
def ask(prompt):
    out = t.llm(
        prompt, 
        max_tokens=1024, 
        stop=["Q:"], 
        temperature=0.0,
        top_p=0.2,
        top_k=10,
        repeat_penalty=1.2,
        stream=True,
    )
    for c in out:
        print(c["choices"][0]["text"], end='')
    print()
    
def ask_with_rag(question, context):
    full_prompt = "Answer question using the context provided below.\n" + \
        f" Context {context}\n\n" + \
        f"Q: {question}\n" + \
        "A:"
    ask(full_prompt)
```
